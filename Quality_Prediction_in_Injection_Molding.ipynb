```python
# Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest, mutual_info_classif, f_classif
from sklearn.model_selection import train_test_split

# Load your dataset into a DataFrame
# Assuming 'df' is your DataFrame and 'quality' is the target variable
X = df.drop(columns=['quality'])
y = df['quality']

# Splitting dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 1. Feature Selection using Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Getting feature importances from the Random Forest model
importances = rf.feature_importances_
feature_importances = pd.Series(importances, index=X.columns)
feature_importances = feature_importances.sort_values(ascending=False)

print("Random Forest Feature Importances:")
print(feature_importances)

# Plotting feature importances from Random Forest
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importances.values, y=feature_importances.index, palette='viridis')
plt.title('Feature Importances using Random Forest')
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.show()

# 2. Feature Selection using Mutual Information
mutual_info_selector = SelectKBest(score_func=mutual_info_classif, k='all')
mutual_info_selector.fit(X_train, y_train)
mutual_info_scores = mutual_info_selector.scores_
mutual_info = pd.Series(mutual_info_scores, index=X.columns)
mutual_info = mutual_info.sort_values(ascending=False)

print("\nMutual Information Scores:")
print(mutual_info)

# Plotting Mutual Information Scores
plt.figure(figsize=(10, 6))
sns.barplot(x=mutual_info.values, y=mutual_info.index, palette='viridis')
plt.title('Mutual Information Scores for Features')
plt.xlabel('Score')
plt.ylabel('Features')
plt.show()

# 3. Feature Selection using ANOVA-F Method
anova_selector = SelectKBest(score_func=f_classif, k='all')
anova_selector.fit(X_train, y_train)
anova_scores = anova_selector.scores_
anova = pd.Series(anova_scores, index=X.columns)
anova = anova.sort_values(ascending=False)

print("\nANOVA F-Scores:")
print(anova)

# Plotting ANOVA F-Scores
plt.figure(figsize=(10, 6))
sns.barplot(x=anova.values, y=anova.index, palette='viridis')
plt.title('ANOVA F-Scores for Features')
plt.xlabel('Score')
plt.ylabel('Features')
plt.show()
```

```python
# Importing required libraries
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.metrics import classification_report, accuracy_score

# Load your dataset into a DataFrame
# Assuming 'df' is your DataFrame and 'quality' is the target variable
X = df.drop(columns=['quality'])
y = df['quality']

# Splitting dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Defining the Logistic Regression model
log_reg = LogisticRegression(solver='liblinear', random_state=42)

# Setting up the hyperparameter grid for RandomizedSearchCV
param_distributions = {
    'C': [0.01, 0.1, 1, 10, 100],  # Regularization strength
    'penalty': ['l1', 'l2'],  # Regularization type
    'solver': ['liblinear', 'saga'],  # Solver to use
    'max_iter': [100, 200, 300, 500]  # Maximum number of iterations
}

# Setting up RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=log_reg,
    param_distributions=param_distributions,
    n_iter=10,  # Number of different combinations to try
    scoring='accuracy',
    cv=5,  # Cross-validation folds
    random_state=42,
    n_jobs=-1  # Use all available cores
)

# Fitting the model with the best hyperparameters
random_search.fit(X_train, y_train)

# Extracting the best model
best_log_reg = random_search.best_estimator_
print("Best Hyperparameters:")
print(random_search.best_params_)

# Making predictions on the test set
y_pred = best_log_reg.predict(X_test)

# Evaluating the model
print("\nAccuracy Score:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Importing required libraries
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import classification_report, accuracy_score

# Load your dataset into a DataFrame
# Assuming 'df' is your DataFrame and 'quality' is the target variable
X = df.drop(columns=['quality'])
y = df['quality']

# Splitting dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Defining the Logistic Regression model
log_reg = LogisticRegression(solver='liblinear', random_state=42)

# Setting up the hyperparameter grid for GridSearchCV
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],  # Regularization strength
    'penalty': ['l1', 'l2'],  # Regularization type
    'solver': ['liblinear', 'saga'],  # Solver to use
    'max_iter': [100, 200, 300, 500]  # Maximum number of iterations
}

# Setting up GridSearchCV
grid_search = GridSearchCV(
    estimator=log_reg,
    param_grid=param_grid,
    scoring='accuracy',
    cv=5,  # Cross-validation folds
    n_jobs=-1  # Use all available cores
)

# Fitting the model with the best hyperparameters
grid_search.fit(X_train, y_train)

# Extracting the best model
best_log_reg = grid_search.best_estimator_
print("Best Hyperparameters:")
print(grid_search.best_params_)

# Making predictions on the test set
y_pred = best_log_reg.predict(X_test)

# Evaluating the model
print("\nAccuracy Score:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Importing required libraries
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import classification_report, accuracy_score
import numpy as np

# Load your dataset into a DataFrame
# Assuming 'df' is your DataFrame and 'quality' is the target variable
X = df.drop(columns=['quality'])
y = df['quality']

# Splitting dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Defining the Logistic Regression model
log_reg = LogisticRegression(solver='liblinear', random_state=42)

# Performing k-Fold Cross Validation with k=5
cv_scores = cross_val_score(log_reg, X_train, y_train, cv=5, scoring='accuracy')

# Printing cross-validation scores and mean accuracy
print("Cross-Validation Scores:", cv_scores)
print("Mean Cross-Validation Accuracy:", np.mean(cv_scores))

# Fitting the model on the training set
log_reg.fit(X_train, y_train)

# Making predictions on the test set
y_pred = log_reg.predict(X_test)

# Evaluating the model
print("\nAccuracy Score on Test Set:", accuracy_score(y_test, y_pred))
print("\nClassification Report on Test Set:")
print(classification_report(y_test, y_pred))
```

```python
# Importing required libraries
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import classification_report, accuracy_score
import numpy as np

# Load your dataset into a DataFrame
# Assuming 'df' is your DataFrame and 'quality' is the target variable
X = df.drop(columns=['quality'])
y = df['quality']

# Splitting dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Defining the Random Forest model
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Performing k-Fold Cross Validation with k=5
cv_scores = cross_val_score(rf, X_train, y_train, cv=3, scoring='accuracy')

# Printing cross-validation scores and mean accuracy
print("Cross-Validation Scores:", cv_scores)
print("Mean Cross-Validation Accuracy:", np.mean(cv_scores))

# Fitting the model on the training set
rf.fit(X_train, y_train)

# Making predictions on the test set
y_pred = rf.predict(X_test)

# Evaluating the model
print("\nAccuracy Score on Test Set:", accuracy_score(y_test, y_pred))
print("\nClassification Report on Test Set:")
print(classification_report(y_test, y_pred))
```
